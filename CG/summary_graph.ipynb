```python
# CG
predictions = {"Yes": 0, "Minor": 0, "No": 24}
issues = { "Redundancy": 11 ,"Fabrication": 5, "MissCall": 22}
import pandas as pd
# create DataFrame
res_df = pd.DataFrame({'Good': [predictions["Yes"], 0],
                   'Minor':[predictions["Minor"], 0],
                   '':[0, predictions["No"]]},
                  index=["Reasonable", "Unreasonable"])
 
 
# create stacked bar chart for monthly temperatures
res_df.plot(kind='bar', stacked=True,  alpha=0.7, width=0.3, figsize=(10,5))

import matplotlib.pyplot as plt
# labels for x & y axis
plt.xticks(rotation=0, fontsize=35)
plt.yticks(fontsize=35)
plt.ylabel('Number', fontsize=35)
plt.legend(fontsize=35) # using a size in points
plt.tight_layout()
plt.savefig('starchat_cg_count_reasonable.pdf')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame.from_records( {"Issues":["Redundancy", "Fabrication", "MissingCall"], "Number":[issues["Redundancy"], issues["Fabrication"], issues["MissCall"]]} )#{"Lack statement tokens": 7, "Lack syntax keywords":9, "Wrong Structure":5}
plt.figure(figsize=(10,5))
sns.barplot(df, x='Issues', y='Number', alpha=0.8, width=0.4, color="steelblue")
plt.xticks(rotation=0, fontsize=25)
plt.yticks(fontsize=35)
plt.xlabel('', fontsize=35)
plt.ylabel('Number', fontsize=35)
plt.tight_layout()
plt.savefig('starchat_cg_issues.pdf')
```
