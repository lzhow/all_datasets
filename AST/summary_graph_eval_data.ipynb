```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulated data for 4 models
predictions = {
    'GPT4': {"Good": 67, "Minor": 4, "No": 4},
    'GPT3.5': {"Good": 49, "Minor": 15, "No": 11},
    'CodeLlama': {"Good": 55, "Minor": 4, "No": 16},
    'StarChat': {"Good": 41, "Minor": 2, "No": 32}
}

issues = {
    'GPT4': {"Lack statement tokens": 0, "Lack syntax keywords": 0, "Wrong Structure": 4},
    'GPT3.5': {"Lack statement tokens": 7, "Lack syntax keywords": 9, "Wrong Structure": 5},
    'CodeLlama': {"Lack statement tokens": 6, "Lack syntax keywords": 3, "Wrong Structure": 7},
    'StarChat': {"Lack statement tokens": 9, "Lack syntax keywords": 8, "Wrong Structure": 24}
}
```

```python
fig, ax = plt.subplots(figsize=(18, 9))

# Plot "Good" and "Minor" side by side in the same column with different colors, and "No" in a separate column
bar_width = 0.3  # Width of the bars
index = np.arange(len(predictions))  # The label locations

# Bar locations for each group
bar1 = ax.bar(index - bar_width/2, [pred['Good'] for pred in predictions.values()], bar_width, label='Good', color='#1f77b4', alpha=0.7)
bar2 = ax.bar(index - bar_width/2, [pred['Minor'] for pred in predictions.values()], bar_width, label='Minor', color='#ff7f0e', alpha=0.7, bottom=[pred['Good'] for pred in predictions.values()])
bar3 = ax.bar(index + bar_width/2, [pred['No'] for pred in predictions.values()], bar_width, label='No', color='#2ca02c', alpha=0.7)

# Set titles and labels
# ax.set_title('Code Quality Comparison (Good vs Minor vs No)')
ax.set_xticks(index)
ax.set_xticklabels(predictions.keys(), fontsize=40)
ax.tick_params(axis='y', labelsize=40)
ax.set_ylabel('Number', fontsize=40)  # Y-axis title font size

# Adjusting the legend to reflect the individual categories with font size
ax.legend(fontsize=30)  # Legend font size

plt.tight_layout()

# Save the figure for Code Quality with distinct colors for "Good" and "Minor" in the same column
code_quality_combined_colors_path = 'ast_reasonable.pdf'
plt.savefig(code_quality_combined_colors_path)

code_quality_combined_colors_path
```

```python
import pandas as pd
```

```python
issues_df = pd.DataFrame(issues).T

# Creating the bar chart with a smaller width to reduce spacing between bars
fig, ax = plt.subplots(figsize=(18, 9))
issues_df.plot(kind='bar', ax=ax, alpha=0.7, width=0.7)  # Adjust the 'width' to control bar spacing

ax.set_ylabel('Number', fontsize=40)  # Y-axis title font size

# Set model names to be centered
ax.set_xticklabels(issues_df.index, rotation=0, ha='center', fontsize=40)
ax.tick_params(axis='y', labelsize=40)
ax.legend(fontsize=30)  # Legend font size

# Adjust layout for saving
plt.tight_layout()

# Save the figure to a PDF file
syntax_issues_shared_path = 'ast_issues.pdf'
plt.savefig(syntax_issues_shared_path)
```
