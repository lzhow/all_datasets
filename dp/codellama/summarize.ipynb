```python

import json
import os
for file_dir in sorted(os.listdir()):
  if os.path.isdir(file_dir):
    # contract

    # JSON
    with open(f'{file_dir}/facts_contract_sample_50.json', 'r') as file:
        data = json.load(file)
    

    #  dp  taint 
    for part in data:
        # 'dp'  'taint' 
        folder_name = 'dp' if part == data[0] else 'taint'

        # A
        for i, A in enumerate(part):
          # 
          file_name = f"{file_dir}/contract/{folder_name}/{i}.txt"  # .txt

          # B
          if os.path.exists(file_name):
              with open(file_name, 'r') as file:
                  content = file.read()
                  A.append(content)  # B

    # JSON
    with open(f'{file_dir}/facts_contract_sample_50_dp_codellama.json', 'w') as file:
        json.dump(data[0], file, indent=4)
    with open(f'{file_dir}/facts_contract_sample_50_taint_codellama.json', 'w') as file:
        json.dump(data[1], file, indent=4)

    # function
    
    # JSON
    with open(f'{file_dir}/facts_function_sample_50.json', 'r') as file:
        data = json.load(file)

    #  dp  taint 
    for part in data:
        # 'dp'  'taint' 
        folder_name = 'dp' if part == data[0] else 'taint'

        # A
        for i, A in enumerate(part):
          # 
          file_name = f"{file_dir}/function/{folder_name}/{i}.txt"  # .txt

          # B
          if os.path.exists(file_name):
              with open(file_name, 'r') as file:
                  content = file.read()
                  A.append(content)  # B

    # JSON
    with open(f'{file_dir}/facts_function_sample_50_dp_codellama.json', 'w') as file:
        json.dump(data[0], file, indent=4)
    with open(f'{file_dir}/facts_function_sample_50_taint_codellama.json', 'w') as file:
        json.dump(data[1], file, indent=4)
```
